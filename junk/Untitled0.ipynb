```python
import numpy as np
from sklearn.metrics.pairwise import rbf_kernel
from scipy.stats import gaussian_kde
from scipy.optimize import minimize_scalar

def compute_entropy(X, gamma):
    K = rbf_kernel(X, gamma=gamma)
    values = K[np.triu_indices(K.shape[0], k=1)]
    kde = gaussian_kde(values)

    x = np.linspace(values.min(), values.max(), 1000)
    pdf = kde(x)

    # Calculate entropy
    entropy = -np.sum(pdf * np.log(pdf + 1e-10)) * (x[1] - x[0])
    return entropy

def negative_entropy(gamma):
    return -compute_entropy(X, gamma)

# Example dataset
X = np.random.rand(100, 10)  # Replace with your dataset

# Optimize gamma
result = minimize_scalar(negative_entropy, bounds=(1e-5, 10), method='bounded')

optimal_gamma = result.x
max_entropy = -result.fun

print("Optimal gamma:", optimal_gamma)
print("Max entropy:", max_entropy)
```
