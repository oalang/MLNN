```python
import numpy as np
rng = np.random.Generator(np.random.PCG64(12345))

n = 15000
m = 150
d = 15

Q = rng.standard_normal((n, n))
I = rng.standard_normal((n, n))
C = rng.standard_normal((m, m))
K = rng.standard_normal((m, m))
A = rng.standard_normal((m, m))
Y = rng.standard_normal((n, n))
B = rng.standard_normal((n, m))
D = rng.standard_normal((m, 1))
E = rng.standard_normal((d, m))


O2_path = np.einsum_path('ij,ij->i', Q, I, optimize='optimal')
R2_path = np.einsum_path('ij,ij->i', C, K, optimize='optimal')
N2_path = np.einsum_path('ij,ij->i', E, E, optimize='optimal')
L2_path = np.einsum_path('ij,ij->j', B, Y @ B, optimize='optimal')
L3_path = np.einsum_path('ij,ik,kj->j', B, Y, B, optimize='optimal')
P2_path = np.einsum_path('ij,jk,lk->il', B, A, B, optimize='optimal')
W2_path = np.einsum_path('ij,j,kj->ik', B, D.ravel(), B, optimize='optimal')
S2_path = np.einsum_path('ij,kj,kl,ml->im', B, A, A, B, optimize='optimal')
X2_path = np.einsum_path('ji,jk,kl->il', B, Y, B, optimize='optimal')
Z2_path = np.einsum_path('ij,kj,kl,lm->im', E, B, Y, B, optimize='optimal')
M2_path = np.einsum_path('ij,jk,lk->il', A, C, A, optimize='optimal')
```
