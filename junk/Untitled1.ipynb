```python
import numpy as np
```

```python
m = 100
n = 1000
A = np.random.rand(m,n)
```

```python
N1 = np.trace(A @ A.T)
N2 = np.linalg.norm(A) ** 2
N3 = np.dot(A.ravel(), A.ravel())
N4 = np.sum(np.square(A))
N5 = np.sum(A ** 2)
N1, N2, N3, N4, N5
```

```python
%timeit N1 = np.trace(A @ A.T)
%timeit N2 = np.linalg.norm(A) ** 2
%timeit N3 = np.dot(A.ravel(), A.ravel())
%timeit N4 = np.sum(np.square(A))
%timeit N5 = np.sum(A ** 2)
%timeit N6 = np.sum(A * A)
```

```python
N1 = np.dot((A @ A.T).ravel(), (A @ A.T).ravel())
N2 = np.dot((A.T @ A).ravel(), (A.T @ A).ravel())
N1, N2
```

```python
B = np.random.rand(3,3)
B = B @ B.T
C = np.random.rand(3,3)
C = C @ C.T

N1 = np.trace(B @ C @ B @ C)
N2 = np.dot((C @ B).ravel(), (C @ B).T.ravel())
N1, N2
```

```python
B = np.random.rand(3,3)
C = np.random.rand(3,3)

N1 = np.trace(B @ C)
N2 = np.dot(B.T.ravel(), C.ravel())
N1, N2
```
