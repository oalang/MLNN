```python
import numpy as np
import matplotlib.pyplot as plt
import loss
```

```python
def zero_one(X):
    return np.where(X > 0, 1, 0)
```

```python
relu = loss.ReLU(6)
smooth_relu = loss.SmoothReLU(5)
smooth_relu_2 = loss.SmoothReLU2(4)
smooth_relu_3 = loss.SmoothReLU3(3)
quadratic = loss.Quadratic(2)
logistic = loss.Logistic(1)
sigmoid = loss.Sigmoid(0)
leaky_relu = loss.LeakyReLU(6, .1)
leaky_smooth_relu = loss.LeakySmoothReLU(5, .2)
leaky_smooth_relu_2 = loss.LeakySmoothReLU2(4, .3)
leaky_smooth_relu_3 = loss.LeakySmoothReLU3(3, .4)
leaky_quadratic = loss.LeakyQuadratic(2, .5)
leaky_logistic = loss.LeakyLogistic(1, .6)
leaky_sigmoid = loss.LeakySigmoid(0, .7)
x = np.linspace(-7, 1, 1024, endpoint=True)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))


ax1.plot(x, relu.func(x), label='ReLU(6)')
ax1.plot(x, smooth_relu.func(x), label='SmoothReLU(5)')
ax1.plot(x, smooth_relu_2.func(x), label='SmoothReLU2(4)')
ax1.plot(x, smooth_relu_3.func(x), label='SmoothReLU3(3)')
ax1.plot(x, quadratic.func(x), label='Quadratic(2)')
ax1.plot(x, logistic.func(x), label='Logistic(1)')
ax1.plot(x, sigmoid.func(x), label='Sigmoid(0)')
ax1.plot(x, zero_one(x), label='ZeroOne')
ax1.set_title("loss functions w/ offset")
ax1.legend(loc='upper left')
ax1.set_xlim(-7, 1)
ax1.set_ylim(-1, 4)
ax1.set_aspect('equal')
ax1.grid()

ax2.plot(x, leaky_relu.func(x), label='LeakyReLU(6, .1)')
ax2.plot(x, leaky_smooth_relu.func(x), label='LeakySmoothReLU(5, .2)')
ax2.plot(x, leaky_smooth_relu_2.func(x), label='LeakySmoothReLU2(4, .3)')
ax2.plot(x, leaky_smooth_relu_3.func(x), label='LeakySmoothReLU3(3, .4)')
ax2.plot(x, leaky_quadratic.func(x), label='LeakyQuadratic(2, .5)')
ax2.plot(x, leaky_logistic.func(x), label='LeakyLogistic(1, .6)')
ax2.plot(x, leaky_sigmoid.func(x), label='LeakySigmoid(0, .7)')
ax2.plot(x, zero_one(x), label='ZeroOne')
ax2.set_title("leaky loss functions w/ offset")
ax2.legend(loc='upper left')
ax2.set_xlim(-7, 1)
ax2.set_ylim(-1, 4)
ax2.set_aspect('equal')
ax2.grid()

ax3.plot(x, relu.grad(x), label='ReLU(6)')
ax3.plot(x, smooth_relu.grad(x), label='SmoothReLU(5)')
ax3.plot(x, smooth_relu_2.grad(x), label='SmoothReLU2(4)')
ax3.plot(x, smooth_relu_3.grad(x), label='SmoothReLU3(3)')
ax3.plot(x, quadratic.grad(x), label='Quadratic(2)')
ax3.plot(x, logistic.grad(x), label='Logistic(1)')
ax3.plot(x, sigmoid.grad(x), label='Sigmoid(0)')
ax3.set_title("gradient functions w/ offset")
ax3.legend(loc='upper left')
ax3.set_xlim(-7, 1)
ax3.set_ylim(-1, 4)
ax3.set_aspect('equal')
ax3.grid()

ax4.plot(x, leaky_relu.grad(x), label='LeakyReLU(6, .1)')
ax4.plot(x, leaky_smooth_relu.grad(x), label='LeakySmoothReLU(5, .2)')
ax4.plot(x, leaky_smooth_relu_2.grad(x), label='LeakySmoothReLU2(5, .3)')
ax4.plot(x, leaky_smooth_relu_3.grad(x), label='LeakySmoothReLU3(3, .4)')
ax4.plot(x, leaky_quadratic.grad(x), label='LeakyQuadratic(2, .5)')
ax4.plot(x, leaky_logistic.grad(x), label='LeakyLogistic(1, .6)')
ax4.plot(x, leaky_sigmoid.grad(x), label='LeakySigmoid(0, .7)')
ax4.set_title("leaky gradient functions w/ offset")
ax4.legend(loc='upper left')
ax4.set_xlim(-7, 1)
ax4.set_ylim(-1, 4)
ax4.set_aspect('equal')
ax4.grid()

fig.tight_layout()
plt.show()
```
