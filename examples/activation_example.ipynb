```python
import numpy as np
import matplotlib.pyplot as plt
import activation
```

```python
def zero_one(X):
    return np.where(X > 0, 1, 0)
```

```python
relu = activation.ReLU(6)
smooth_relu_1 = activation.SmoothReLU1(5)
smooth_relu_2 = activation.SmoothReLU2(4)
smooth_relu_3 = activation.SmoothReLU3(3)
logistic = activation.Logistic(2)
selu = activation.SELU(1)
sigmoid = activation.Sigmoid(0)

leaky_relu = activation.LeakyReLU(6, .1)
leaky_smooth_relu_1 = activation.LeakySmoothReLU1(5, .2)
leaky_smooth_relu_2 = activation.LeakySmoothReLU2(4, .3)
leaky_smooth_relu_3 = activation.LeakySmoothReLU3(3, .4)
leaky_logistic = activation.LeakyLogistic(2, .5)
leaky_selu = activation.LeakySELU(1, .6)
leaky_sigmoid = activation.LeakySigmoid(0, .7)

x = np.linspace(-7, 1, 1024, endpoint=True)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))

ax1.plot(x, relu.func(x), label='ReLU(6)')
ax1.plot(x, smooth_relu_1.func(x), label='SmoothReLU1(5)')
ax1.plot(x, smooth_relu_2.func(x), label='SmoothReLU2(4)')
ax1.plot(x, smooth_relu_3.func(x), label='SmoothReLU3(3)')
ax1.plot(x, logistic.func(x), label='Logistic(2)')
ax1.plot(x, selu.func(x), label='SELU(1)')
ax1.plot(x, sigmoid.func(x), label='Sigmoid(0)')
ax1.plot(x, zero_one(x), label='ZeroOne')
ax1.set_title("loss functions w/ offset")
ax1.legend(loc='upper left')
ax1.set_xlim(-7, 1)
ax1.set_ylim(-1, 4)
ax1.set_aspect('equal')
ax1.grid()

ax2.plot(x, leaky_relu.func(x), label='LeakyReLU(6, .1)')
ax2.plot(x, leaky_smooth_relu_1.func(x), label='LeakySmoothReLU1(5, .2)')
ax2.plot(x, leaky_smooth_relu_2.func(x), label='LeakySmoothReLU2(4, .3)')
ax2.plot(x, leaky_smooth_relu_3.func(x), label='LeakySmoothReLU3(3, .4)')
ax2.plot(x, leaky_logistic.func(x), label='LeakyLogistic(2, .5)')
ax2.plot(x, leaky_selu.func(x), label='LeakySELU(1, .6)')
ax2.plot(x, leaky_sigmoid.func(x), label='LeakySigmoid(0, .7)')
ax2.plot(x, zero_one(x), label='ZeroOne')
ax2.set_title("leaky loss functions w/ offset")
ax2.legend(loc='upper left')
ax2.set_xlim(-7, 1)
ax2.set_ylim(-1, 4)
ax2.set_aspect('equal')
ax2.grid()

ax3.plot(x, relu.grad(x), label='ReLU(6)')
ax3.plot(x, smooth_relu_1.grad(x), label='SmoothReLU1(5)')
ax3.plot(x, smooth_relu_2.grad(x), label='SmoothReLU2(4)')
ax3.plot(x, smooth_relu_3.grad(x), label='SmoothReLU3(3)')
ax3.plot(x, logistic.grad(x), label='Logistic(2)')
ax3.plot(x, selu.grad(x), label='SELU(1)')
ax3.plot(x, sigmoid.grad(x), label='Sigmoid(0)')
ax3.set_title("gradient functions w/ offset")
ax3.legend(loc='upper left')
ax3.set_xlim(-7, 1)
ax3.set_ylim(-1, 4)
ax3.set_aspect('equal')
ax3.grid()

ax4.plot(x, leaky_relu.grad(x), label='LeakyReLU(6, .1)')
ax4.plot(x, leaky_smooth_relu_1.grad(x), label='LeakySmoothReLU1(5, .2)')
ax4.plot(x, leaky_smooth_relu_2.grad(x), label='LeakySmoothReLU2(5, .3)')
ax4.plot(x, leaky_smooth_relu_3.grad(x), label='LeakySmoothReLU3(3, .4)')
ax4.plot(x, leaky_logistic.grad(x), label='LeakyLogistic(2, .5)')
ax4.plot(x, leaky_selu.grad(x), label='LeakySELU(1, .6)')
ax4.plot(x, leaky_sigmoid.grad(x), label='LeakySigmoid(0, .7)')
ax4.set_title("leaky gradient functions w/ offset")
ax4.legend(loc='upper left')
ax4.set_xlim(-7, 1)
ax4.set_ylim(-1, 4)
ax4.set_aspect('equal')
ax4.grid()

fig.tight_layout()
plt.show()
```

```python
relu = activation.ReLU(1)
smooth_relu_1 = activation.SmoothReLU1(1)
smooth_relu_2 = activation.SmoothReLU2(1)
smooth_relu_3 = activation.SmoothReLU3(1)
logistic = activation.Logistic(1)
selu = activation.SELU(1)
sigmoid = activation.Sigmoid(0)

leaky_relu = activation.LeakyReLU(1, .25)
leaky_smooth_relu_1 = activation.LeakySmoothReLU1(1, .25)
leaky_smooth_relu_2 = activation.LeakySmoothReLU2(1, .25)
leaky_smooth_relu_3 = activation.LeakySmoothReLU3(1, .25)
leaky_logistic = activation.LeakyLogistic(1, .25)
leaky_selu = activation.LeakySELU(1, .25)
leaky_sigmoid = activation.LeakySigmoid(0, .25)

x = np.linspace(-2, 1, 1024, endpoint=True)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 10))

ax1.plot(x, relu.func(x), label='ReLU(1)')
ax1.plot(x, smooth_relu_1.func(x), label='SmoothReLU1(1)')
ax1.plot(x, smooth_relu_2.func(x), label='SmoothReLU2(1)')
ax1.plot(x, smooth_relu_3.func(x), label='SmoothReLU3(1)')
ax1.plot(x, logistic.func(x), label='Logistic(1)')
ax1.plot(x, selu.func(x), label='SELU(1)')
ax1.plot(x, sigmoid.func(x), label='Sigmoid(0)')
ax1.plot(x, zero_one(x), label='ZeroOne')
ax1.set_title("loss functions w/ offset")
ax1.legend(loc='upper left')
ax1.set_xlim(-2, 1)
ax1.set_ylim(-.5, 1.5)
ax1.set_aspect('equal')
ax1.set_xticks(np.arange(-2, 1.5, .5))
ax1.set_yticks(np.arange(-.5, 2, .5))
ax1.grid()

ax2.plot(x, leaky_relu.func(x), label='LeakyReLU(1, .25)')
ax2.plot(x, leaky_smooth_relu_1.func(x), label='LeakySmoothReLU1(1, .25)')
ax2.plot(x, leaky_smooth_relu_2.func(x), label='LeakySmoothReLU2(1, .25)')
ax2.plot(x, leaky_smooth_relu_3.func(x), label='LeakySmoothReLU3(1, .25)')
ax2.plot(x, leaky_logistic.func(x), label='LeakyLogistic(1, .25)')
ax2.plot(x, leaky_selu.func(x), label='LeakySELU(1, .25)')
ax2.plot(x, leaky_sigmoid.func(x), label='LeakySigmoid(0, .25)')
ax2.plot(x, zero_one(x), label='ZeroOne')
ax2.set_title("leaky loss functions w/ offset")
ax2.legend(loc='upper left')
ax2.set_xlim(-2, 1)
ax2.set_ylim(-.5, 1.5)
ax2.set_aspect('equal')
ax2.set_xticks(np.arange(-2, 1.5, .5))
ax2.set_yticks(np.arange(-.5, 2, .5))
ax2.grid()

ax3.plot(x, relu.grad(x), label='ReLU(1)')
ax3.plot(x, smooth_relu_1.grad(x), label='SmoothReLU1(1)')
ax3.plot(x, smooth_relu_2.grad(x), label='SmoothReLU2(1)')
ax3.plot(x, smooth_relu_3.grad(x), label='SmoothReLU3(1)')
ax3.plot(x, logistic.grad(x), label='Logistic(1)')
ax3.plot(x, selu.grad(x), label='SELU(1)')
ax3.plot(x, sigmoid.grad(x), label='Sigmoid(0)')
ax3.set_title("gradient functions w/ offset")
ax3.legend(loc='upper left')
ax3.set_xlim(-2, 1)
ax3.set_ylim(-.5, 1.5)
ax3.set_aspect('equal')
ax3.set_xticks(np.arange(-2, 1.5, .5))
ax3.set_yticks(np.arange(-.5, 2, .5))
ax3.grid()

ax4.plot(x, leaky_relu.grad(x), label='ReLU(1, .25)')
ax4.plot(x, leaky_smooth_relu_1.grad(x), label='LeakySmoothReLU1(1, .25)')
ax4.plot(x, leaky_smooth_relu_2.grad(x), label='LeakySmoothReLU2(1, .25)')
ax4.plot(x, leaky_smooth_relu_3.grad(x), label='LeakySmoothReLU3(1, .25)')
ax4.plot(x, leaky_logistic.grad(x), label='LeakyLogistic(1, .25)')
ax4.plot(x, leaky_selu.grad(x), label='LeakySELU(1, .25)')
ax4.plot(x, leaky_sigmoid.grad(x), label='LeakySigmoid(0, .25)')
ax4.set_title("leakey gradient functions w/ offset")
ax4.legend(loc='upper left')
ax4.set_xlim(-2, 1)
ax4.set_ylim(-.5, 1.5)
ax4.set_aspect('equal')
ax4.set_xticks(np.arange(-2, 1.5, .5))
ax4.set_yticks(np.arange(-.5, 2, .5))
ax4.grid()

fig.tight_layout()
plt.show()
```
